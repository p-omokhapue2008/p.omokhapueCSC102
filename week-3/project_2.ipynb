```python


def solve_quadratic():
    print("\nQuadratic Equation: Ax^2 + Bx + C = 0")
    
    A = float(input("Enter coefficient A: "))
    B = float(input("Enter coefficient B: "))
    C = float(input("Enter coefficient C: "))

    D = B**2 - 4*A*C  # Discriminant

    if D >= 0:
        root1 = (-B + math.sqrt(D)) / (2 * A)
        root2 = (-B - math.sqrt(D)) / (2 * A)
        print("Roots:", root1, "and", root2)
    else:
        root1 = (-B + cmath.sqrt(D)) / (2 * A)
        root2 = (-B - cmath.sqrt(D)) / (2 * A)
        print("Complex Roots:", root1, "and", root2)

def solve_cubic():
    print("\nCubic Equation: Ax^3 + Bx^2 + Cx + D = 0")

    A = float(input("Enter coefficient A: "))
    B = float(input("Enter coefficient B: "))
    C = float(input("Enter coefficient C: "))
    D = float(input("Enter coefficient D: "))

    # Guessing method (beginner-friendly)
    for x in range(-100, 101):  # Check integer roots
        if A*x**3 + B*x**2 + C*x + D == 0:
            print("One root is:", x)
            return
    print("No simple integer root found.")

def solve_quartic():
    print("\nQuartic Equation: Ax^4 + Bx^3 + Cx^2 + Dx + E = 0")

    A = float(input("Enter coefficient A: "))
    B = float(input("Enter coefficient B: "))
    C = float(input("Enter coefficient C: "))
    D = float(input("Enter coefficient D: "))
    E = float(input("Enter coefficient E: "))

    # Checking integer roots (beginner-friendly)
    for x in range(-100, 101):
        if A*x**4 + B*x**3 + C*x**2 + D*x + E == 0:
            print("One root is:", x)
            return
    print("No simple integer root found.")

# Menu for user to choose operation
while True:
    print("\nChoose an operation:")
    print("1 - Solve a Quadratic Equation")
    print("2 - Solve a Cubic Equation")
    print("3 - Solve a Quartic Equation")
    print("4 - Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == "1":
        solve_quadratic()
    elif choice == "2":
        solve_cubic()
    elif choice == "3":
        solve_quartic()
    elif choice == "4":
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please select a valid option.")

  
```
